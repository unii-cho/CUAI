### kmeans clustering 알고리즘
1. prepare data
2. decide how many clusters you need
3. choose initial center of cluster(centroid)
-> randomly select centroid
-> manually assign centroid(인위적으로 지정)
그래도 불만족스러울 시
-> kmean++ initial centroid(첫번째 데이터포인트 위치에 센트로이드 놓기,
두번째는 첫번째 데이터포인트에서 가장 먼 위치,
세번째는 첫번째와 두번째에서 공통적으로 가장 먼 위치)
4. assign data point to nearest cluster
5. move centroid to the center of its cluster
6. repeat step 4 and step 5 util there is no assigned cluster change


```python
df=pd.DataFrame(columns=['x','y'])
```

```python
df.loc[0]=[3,1]
df.loc[1]=[4,1]
df.loc[2]=[3,2]
df.loc[3]=[4,2]
df.loc[4]=[10,5]
df.loc[5]=[10,6]
df.loc[6]=[11,5]
df.loc[7]=[11,6]
df.loc[8]=[15,1]
df.loc[9]=[15,2]
df.loc[10]=[16,1]
df.loc[11]=[16,2]
```

```python
df.head(20)
```

```python
import seaborn as sns
#visualize data point
sns.lmplot('x','y',data=df,fit_reg=False, scatter_kws={"s": 200})
#x-axis, y-axis, data, no line, marker size

#title
plt.title('kmean plot')

#x-axis label
plt.xlabel('x')

#y-axis label
plt.ylabel('y')
```

### k-mean clustering

```python
# convert dataframe to numpy array
data_points = df.values
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3).fit(data_points)
#inital centroid를 넣지 않으면 k mean++로 돼 
```

```python
#cluster id for each data point
kmeans.labels_
```

```python
# this is final centroids position
kmeans.cluster_centers_
```

```python
df['cluster_id']=kmeans.labels_
```

```python
sns.lmplot('x','y',data=df,fit_reg=False, scatter_kws={"s":150},hue="cluster_id")
#x-axis, y-axis, data, no line  , marker size,   color
#title
plt.title('after kmean clustering')
```
